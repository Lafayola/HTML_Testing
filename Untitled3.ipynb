```python
# Start Spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("StopWords").getOrCreate()
```

```python
from pyspark.ml.feature import Tokenizer, StopWordsRemover
```

```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python
import io
df2 = pd.read_csv(io.BytesIO(uploaded['Fake.csv']))
```

```python
# Read in data from S3 Buckets
from pyspark import SparkFiles
url ="Fake.csv"
spark.sparkContext.addFile(url)
df = spark.read.csv(SparkFiles.get("Fake.csv"), sep=",", header=True)

# Show DataFrame
df.show(truncate=False)
```

```python
# Tokenize DataFrame
review_data = Tokenizer(inputCol="text", outputCol="Words")
```

```python
# Transform DataFrame
reviewed = review_data.transform(df)
reviewed.show(truncate=False)
```

```python
# Remove stop words
remover = StopWordsRemover(inputCol="Words", outputCol="filtered")
```

```python
# Transform new DataFrame
newFrame = remover.transform(reviewed)
newFrame.show(truncate=False)
```

```python
# Show simplified review
newFrame.select("filtered").show(truncate=False)
```
